```python
import cadquery as cq

# Define the dimensions of the LEGO brick
length = 16.0  # mm
width = 8.0  # mm
height = 9.6  # mm
stud_diameter = 4.8  # mm
stud_height = 1.8  # mm

plane = cq.Workplane("XY")

# Create the first 1x2 LEGO brick
brick = plane.box(length, width, height)
stud = plane.workplane(offset=height-5).circle(stud_diameter / 2).extrude(stud_height).translate((-3.5,0,0))
brick = brick.union(stud)
stud = plane.workplane(offset=height-5).circle(stud_diameter / 2).extrude(stud_height).translate((+3.5,0,0))
brick = brick.union(stud)

assembly = (
    Assembly(brick, color=Color("Red"))
    .add(brick.rotate((0, 0, 0), (0, 0, 1), 90), loc=Location(Vector(length-4,width/2,0)), color=Color("Red"))
    .add(brick, loc=Location(Vector(length/2,0,height)), color=Color("Green"))
)

# Display the assembly
display(assembly)
```
