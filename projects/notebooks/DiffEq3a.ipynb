# Diff Eq - second order

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sympy import symbols, solve, simplify, factor, roots, Function, Eq, Derivative, sin

pd.set_option("display.max.columns", None)
pd.set_option("display.precision", 2)
```

The differential equation:
$$
\frac{d^2}{dt^2} f(t) + 100 f(t) = 0
$$

The initial condition:
$$
f(0) = 0
$$

```python
# Suppose we want to plot this equation from time = 0 to 10
tStart = 0
tStop = 2

t = tStart
dt = 0.01
f = 0
df = 1

# For plotting
F = []
T = []

print(f"""
     i       t          f
------ ------- ----------
""")
for i in range(int((tStop - tStart)/dt)):
    print(f"{i:>6} {t:>7.2f} {f:>10.2f}")
    F.append(f)
    T.append(t)
    
    # Get the next value...
    df = df - 100.0 * f * dt
    
    t = t + dt
    f = f + df * dt
```

```python
plt.plot(T, F, 'b', label="f")
plt.legend()
plt.show()
```

```python
# Find the actual solution analytically.
# Here's how sympy would solve the equation if it was a calculus student in college:

from sympy import symbols, solve, dsolve, simplify, factor, roots, Function, Eq, Derivative, sin

f,t = symbols('f,t')

# Define the unknown function
f = Function('f')

# Define the differential equation
diffeq = Eq(f(t).diff(t, t) + 100*f(t), 0)

# Define initial conditions
ic = {f(0): 0, f(t).subs(t, 0): 0}

# Solve the differential equation
sol = dsolve(diffeq, ics=ic)

print("The differential equation:")
diffeq
```
