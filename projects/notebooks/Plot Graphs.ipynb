```python
from sympy import symbols, solve, simplify, factor, roots
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create an array of x values from -10 to 4π (100 values)
x = np.linspace(-10, 4 * np.pi, 100)

# Plot the three curves each with a specific color
plt.plot(x, np.sin(x), 'r', label="Sin")
plt.plot(x, np.cos(x), 'g', label="Cos")
plt.plot(x, 0.1*x, 'b', label="Just a line with a slope")

# Add a legend at the top
plt.legend(bbox_to_anchor=(0.5, 1.05), loc='lower center')

# Show it
plt.show()
```
